```python
# Steps to be followed...

# 1. Import libraries
import matplotlib.pyplot as plt

# 2. Prepare data
x = [10,20,30,40,50]
y = [10,20,30,40,50]

# 3. Set-up plot
fig, ax = plt.subplots(figsize=(5,5))    # (width, height)

# 4. Plot data
ax.plot(x,y)

# 5. Customize plot
ax.set(title = "Simple graph",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 6. Saving the figure (optional)
# fig.savefig("PATH")
```

# SCATTER-PLOT

```python
import numpy as np
x = np.linspace(0,10,100)
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))    # sin(x)
```

# VERTICAL BAR-GRAPH

```python
# Creating graphs from dictionaries
nut_butter_prices = {"Almond Butter" : 10,
                     "Penut Butter" : 8,
                     "Cashew Butter" : 12}

fix, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = "Nut butter Prices",
       ylabel = "Prices")
```

# HORIZONTAL BAR-GRAPH

```python
# Horizontal histogram
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title = "Nut butter Prices",
       xlabel = "Prices")
```

# HISTOGRAM

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

# PLOT MULTIPLE GRAPHS IN SINGLE FIGURE

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(15,10))

ax1.plot(x, x/2)
ax1.set(title="Graph")

ax2.scatter(np.random.random(10), np.random.random(10))
ax2.set(title="Scatter")

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.set(title="Bar Graph")

ax4.hist(x)
ax4.set(title="Histogram")
```
